# <center> Filtrado de delitos por horarios </center>

Siguiendo un heatmap presentado en el EDA donde se mostraba la tasa de delitos por horario, se seleccionaron solo los datos correspondidos a los horarios donde la tasa se considera alta.

```python
import pandas as pd
import numpy as np
```

```python
FGJ_carpetas = pd.read_csv("../Data/FGJ_carpetas_homo.csv", parse_dates=["fecha_hechos", "fecha_inicio"])
print(FGJ_carpetas.info())
```

```python
FGJ_carpetas.index = pd.DatetimeIndex(FGJ_carpetas.fecha_hechos)
FGJ_carpetas.head(3)
```

```python
FGJ_carpetas["categoria_delito_homo"].value_counts()
```

Retiramos:
- Delitos de bajo impacto
- Hecho no delictivo

```python
robo_obj = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO DE OBJETOS"].between_time("07:00", "16:00").sort_index()
vio_fam = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "VIOLENCIA FAMILIAR"].between_time("10:00", "23:59").sort_index()
robo_trans = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == 
                          "ROBO A TRANSEUNTE EN VIA PUBLICA CON Y SIN VIOLENCIA"].between_time("11:00", "23:00").sort_index()
fraude = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "FRAUDE"].between_time("10:00", "16:00").sort_index()
robo_neg_sin_vio = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A NEGOCIO SIN VIOLENCIA"].between_time("09:00", "22:00").sort_index()
amenazas = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "AMENAZAS"].between_time("10:00", "22:00").sort_index()
robo_car = pd.concat([FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO DE VEHICULO CON Y SIN VIOLENCIA"].between_time("18:00", "23:59"), 
          FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO DE VEHICULO CON Y SIN VIOLENCIA"].between_time("06:00", "08:59")]).sort_index()
robo_casa_sin = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A CASA SIN VIOLENCIA"].between_time("10:00", "21:59").sort_index()
robo_pasa = pd.concat([FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A PASAJERO A BORDO DE OTROS TRANSPORTE CON Y SIN VIOLENCIA"].between_time("07:00", "10:00"), 
                       FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A PASAJERO A BORDO DE OTROS TRANSPORTE CON Y SIN VIOLENCIA"].between_time("14:00", "21:59")]).sort_index()
sexuales = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "OTROS DELITOS DE INDOLE SEXUAL"].between_time("09:00", "20:59").sort_index()
robo_neg_con = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A NEGOCIO CON VIOLENCIA"].between_time("15:00", "21:59").sort_index()
robo_rep = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A REPARTIDOR CON Y SIN VIOLENCIA"].between_time("09:00", "17:59").sort_index()
robo_metro = pd.concat([FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A PASAJERO A BORDO DEL METRO CON Y SIN VIOLENCIA"].between_time("07:00", "09:59"),
                        FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "ROBO A PASAJERO A BORDO DEL METRO CON Y SIN VIOLENCIA"].between_time("14:00", "20:59")]).sort_index()
homi = pd.concat([FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "HOMICIDIO DOLOSO"].between_time("00:00", "02:59"), 
                  FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "HOMICIDIO DOLOSO"].between_time("19:00", "23:59")]).sort_index()
viola = FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "VIOLACION"].between_time("19:00", "23:59").sort_index()
lesiones = pd.concat([FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "LESIONES DOLOSAS POR DISPARO DE ARMA DE FUEGO"].between_time("00:00", "01:59"),
                      FGJ_carpetas[FGJ_carpetas["categoria_delito_homo"] == "LESIONES DOLOSAS POR DISPARO DE ARMA DE FUEGO"].between_time("18:00", "23:59")]).sort_index()
```

Guardamos los datos de cada delito en archivos distintos

```python
robo_obj.to_csv("../Data/robo_objetos.csv")
vio_fam.to_csv("../Data/violencia_familiar.csv")
robo_trans.to_csv("../Data/robo_transeuntes.csv")
fraude.to_csv("../Data/fraude.csv")
robo_neg_sin_vio.to_csv("../Data/robo_neg_sin.csv")
amenazas.to_csv("../Data/amenazas.csv")
robo_car.to_csv("../Data/robo_carros.csv")
robo_casa_sin.to_csv("../Data/robo_casa_sin.csv")
robo_pasa.to_csv("../Data/robo_pasajeros.csv")
sexuales.to_csv("../Data/sexuales.csv")
robo_neg_con.to_csv("../Data/robo_neg_con.csv")
robo_rep.to_csv("../Data/robo_repartidores.csv")
robo_metro.to_csv("../Data/robo_metro.csv")
homi.to_csv("../Data/homicidios.csv")
viola.to_csv("../Data/violaciones.csv")
lesiones.to_csv("../Data/lesiones_arma.csv")
```
